```python
import os
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

folder_path = 'F:/singapore/ResaleFlatPrices'

csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

dataframes = [pd.read_csv(os.path.join(folder_path, f)) for f in csv_files]

all_data = pd.concat(dataframes, ignore_index=True)
all_data
```

```python
all_data.shape
```

```python
## Handling Null Values
```

```python
all_data.isnull().sum()
```

```python
#dropping null remaining_lease values
all_data = all_data.dropna()

all_data.info()
```

```python
all_data
```

```python
all_data.drop_duplicates(subset='street_name', inplace=True)
all_data['month'].value_counts()
all_data.tail(10)
```

```python


# Function to clean 'Story Range', remove 'TO' and add the two numbers
def clean_and_add_story_range(story_range):
   
    parts = story_range.replace(' TO ', ' ').split()
    
    
    return (int(parts[0]) + int(parts[1]))/2


all_data['storey_range'] = all_data['storey_range'].apply(clean_and_add_story_range)
```

```python
all_data['flat_type'].value_counts()
```

```python

all_data = all_data[~all_data['flat_type'].isin(['EXECUTIVE', 'MULTI-GENERATION'])]
all_data['flat_type'] = all_data['flat_type'].str.extract('(\d+)', expand=False)


all_data['flat_type'].value_counts()
```

```python
all_data['resale_price'] = all_data['resale_price'].astype('float')
all_data['floor_area_sqm'] = all_data['floor_area_sqm'].astype('float')
all_data['lease_commence_date'] = all_data['lease_commence_date'].astype('int64')
all_data['flat_type'] = all_data['flat_type'].astype('float')

```

```python
print(all_data.info())
all_data['town'].value_counts()
```

```python
all_data['remaining_lease'].unique
```

```python
import re


def extract_integers(description):
    
    description = str(description)
    
    integers = re.findall(r'\d+', description)
    if integers:
        return int(integers[0])
    return None 


all_data['remaining_lease'] = all_data['remaining_lease'].apply(extract_integers)

all_data['remaining_lease'].unique
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


col = ['remaining_lease','floor_area_sqm','storey_range','resale_price']


def plot(df, column):
    plt.figure(figsize=(20,5))
    plt.subplot(1,3,1)
    sns.boxplot(data=df, x=column)
    plt.title(f'Box Plot for {column}')

    plt.subplot(1,3,2)
    sns.histplot(data=df, x=column, kde=True, bins=50)
    plt.title(f'Distribution Plot for {column}')

    plt.subplot(1,3,3)
    sns.violinplot(data=df, x=column)
    plt.title(f'Violin Plot for {column}')
    plt.show()
```

```python
for i in ['remaining_lease','floor_area_sqm','storey_range','resale_price']:
    plot(all_data, i)
```

```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()


scaling_features = all_data[['town_encoded','remaining_lease','floor_area_sqm','storey_range_log','flat_model_encoded']]

X_scaled = scalar.fit_transform(scaling_features)

scaling_features_df = pd.DataFrame(X_scaled, columns=['town_encoded','remaining_lease','floor_area_sqm','storey_range_log','flat_model_encoded'])


```

```python

scaling_features_df
```

```python
### Data Preprocessing
```

```python
X = scaling_features_df
y = all_data['resale_price_log']
X


```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 5)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lr = LinearRegression()
lr.fit (X_train, y_train)
y_pred_lr = lr.predict(X_test)

mse = mean_squared_error(y_test, y_pred_lr)
r2 = r2_score(y_test, y_pred_lr)
print('Mean squared error:', mse)
print('R-squared:', r2)
```

```python
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train) 
y_pred_dtr = dtr.predict(X_test)

mse = mean_squared_error(y_test, y_pred_dtr)
r2 = r2_score(y_test, y_pred_dtr)
print('Mean squared error:', mse)
print('R-squared:', r2)
```

```python
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

mse = mean_squared_error(y_test, y_pred_rf)
r2 = r2_score(y_test, y_pred_rf)
print('Mean squared error:', mse)
print('R-squared:', r2)
```

```python
gbr = GradientBoostingRegressor(random_state=42)
gbr.fit(X_train,y_train)

y_pred_gbr = gbr.predict(X_test)

mse = mean_squared_error(y_test, y_pred_gbr)
r2 = r2_score(y_test, y_pred_gbr)
print('Mean squared error:', mse)
print('R-squared:', r2)
```

```python
best_gbr = GradientBoostingRegressor(
    random_state=42, 
    learning_rate=0.1, 
    max_depth=3, 
    max_features='sqrt', 
    min_samples_leaf=2, 
    min_samples_split=10, 
    n_estimators=200, 
    subsample=0.8
)
best_gbr.fit(X_train,y_train)

y_pred_gbr = best_gbr.predict(X_test)

mse = mean_squared_error(y_test, y_pred_gbr)
r2 = r2_score(y_test, y_pred_gbr)
print('Mean squared error:', mse)
print('R-squared:', r2)
```
